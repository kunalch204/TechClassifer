## Step 1: Preprocessing - Detect Face and Eyes

In this first step, our goal is to identify and isolate faces within an image. Due to the variability in images, from the number of faces to their visibility, detecting a clear face is crucial. Once a face is detected, we aim to locate the eyes. An image is considered suitable for further processing only if it has a detectable face with two eyes.

We will use OpenCV's Haar Cascade classifiers for this purpose. These classifiers are designed for efficient object detection, making them ideal for identifying facial features amidst various backgrounds.

The process involves:
1. **Face Detection**: Employing a Haar Cascade classifier trained for face detection.
2. **Eye Detection**: Applying a Haar Cascade classifier for eye detection within the detected facial region.
3. **Validation**: Retaining images with a single face and two detected eyes.

Next, we will implement this preprocessing step using the appropriate OpenCV modules and pre-trained Haar Cascade classifiers.


```python
img = cv2.imread('./test_images/elon_musk.jpg')
img.shape
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
plt.imshow(gray, cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('./haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('./haarcascades/haarcascade_eye.xml')

faces = face_cascade.detectMultiScale(gray, 1.3, 5)
faces
```

```python
(x,y,w,h) = faces[0]
x,y,w,h
```

```python
face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(face_img)
```

```python
cv2.destroyAllWindows()
for (x,y,w,h) in faces:
    face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        

plt.figure()
plt.imshow(face_img, cmap='gray')
plt.show()
```

# Step 2: Preprocessing: Crop the Facial Region of the Image

## Purpose of Preprocessing
The goal of preprocessing is to automate the extraction of the facial region from a set of images. By doing so, we can store these cropped regions, known as Regions of Interest (ROIs), in a separate folder for further processing. This step is crucial as it allows us to focus on the facial features which are of primary interest in many computer vision tasks.

## Automated Cropping
To streamline the workflow and increase efficiency, we'll implement a function that automatically detects and crops the face from an image. The cropped face will only be saved if the image meets our quality criteria. This approach ensures that we filter out images that might degrade the performance of subsequent processing steps due to quality issues such as low resolution or poor lighting.

## Quality Control
As part of the function, we will include checks to assess the validity of the detected face. Images that do not meet the set quality standards, or where a face is not detected, will not be processed further. This automation saves significant manual effort and allows for the scalability of the face detection system.


```python
%matplotlib inline
plt.imshow(roi_color, cmap='gray')
```

```python
cropped_img = np.array(roi_color)
cropped_img.shape
```

```python
def get_cropped_image_if_2_eyes(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >= 2:
            return roi_color
```

```python
path_to_data = "./dataset/"
path_to_cr_data = "./dataset/cropped/"
```

```python
import os
img_dirs = []
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        img_dirs.append(entry.path)
```

```python
img_dirs
```

```python
import shutil
if os.path.exists(path_to_cr_data):
     shutil.rmtree(path_to_cr_data)
os.mkdir(path_to_cr_data)
```

```python
cropped_image_dirs = []
ceos_file_names_dict = {}
for img_dir in img_dirs:
    count = 1
    ceo_name = img_dir.split('/')[-1]
    ceos_file_names_dict[ceo_name] = []
    for entry in os.scandir(img_dir):
        roi_color = get_cropped_image_if_2_eyes(entry.path)
        if roi_color is not None:
            cropped_folder = path_to_cr_data + ceo_name
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_image_dirs.append(cropped_folder)
                print("Generating cropped images in folder: ",cropped_folder)
            cropped_file_name = ceo_name + str(count) + ".png"
            cropped_file_path = cropped_folder + "/" + cropped_file_name
            cv2.imwrite(cropped_file_path, roi_color)
            ceos_file_names_dict[ceo_name].append(cropped_file_path)
            count += 1
```

# Step 3: Wavelet Transform for Feature Enhancement

Wavelet transform is a powerful tool for image analysis, particularly useful for enhancing features such as edges. In the context of facial feature detection, applying a wavelet transform can significantly improve the visibility of key features like eyes, nose, and lips. This step involves converting images into a wavelet domain, where the edges and contours become more pronounced, aiding in the subsequent feature detection processes.

Below is the implementation of a function `w2d` that applies a wavelet transform to an image. This function converts the image to grayscale, processes it with a specified wavelet (defaulting to 'haar'), and reconstructs the image to highlight its edges and contours.


```python
import numpy as np
import pywt
import cv2    

def w2d(img, mode='haar', level=1):
    # Convert to grayscale
    imArray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Convert to float
    imArray = np.float32(imArray) / 255
    # Compute wavelet coefficients
    coeffs = pywt.wavedec2(imArray, mode, level=level)
    # Zeroing out the approximation coefficients
    coeffs_H = list(coeffs)
    coeffs_H[0] *= 0
    # Reconstructing the image from the modified coefficients
    imArray_H = pywt.waverec2(coeffs_H, mode) * 255
    imArray_H = np.uint8(imArray_H)

    return imArray_H
```

```python
ceo_file_names_dict = {}
for img_dir in cropped_image_dirs:
    ceo_name = img_dir.split('/')[-1]
    file_list = []
    for entry in os.scandir(img_dir):
        file_list.append(entry.path)
    ceo_file_names_dict[ceo_name] = file_list
ceo_file_names_dict
```

```python
class_dict = {}
count = 0
for ceo_name in ceo_file_names_dict.keys():
    class_dict[ceo_name] = count
    count = count + 1
class_dict
```

```python
X, y = [], []
for ceo_name, training_files in ceo_file_names_dict.items():
    for training_image in training_files:
        img = cv2.imread(training_image)
        scalled_raw_img = cv2.resize(img, (32, 32))
        img_har = w2d(img,'db1',5)
        scalled_img_har = cv2.resize(img_har, (32, 32))
        combined_img = np.vstack((scalled_raw_img.reshape(32*32*3,1),scalled_img_har.reshape(32*32,1)))
        X.append(combined_img)
        y.append(class_dict[ceo_name]) 
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

# Model Parameters Overview

Defined parameter grids for hyperparameter tuning of three different classifiers: SVM, Random Forest, and Logistic Regression. Each model's specific parameters are set to explore a range of values during the tuning process to find the optimal configuration.

- **SVM (Support Vector Machine):**
  - `C`: Regularization parameter with values [1, 10, 100, 1000].
  - `kernel`: Specifies the kernel type to be used with options 'rbf' and 'linear'.

- **Random Forest:**
  - `n_estimators`: Number of trees in the forest with options [1, 5, 10].

- **Logistic Regression:**
  - `C`: Inverse of regularization strength with choices [1, 5, 10].

This setup is aimed at optimizing each model's performance through GridSearchCV, allowing a comprehensive search over specified parameter values.


```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto',probability=True),
        'params' : {
            'svc__C': [1,10,100,1000],
            'svc__kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'randomforestclassifier__n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'logisticregression__C': [1,5,10]
        }
    }
}
```

```python
scores = []
best_estimators = {}
import pandas as pd
for algo, mp in model_params.items():
    pipe = make_pipeline(StandardScaler(), mp['model'])
    clf =  GridSearchCV(pipe, mp['params'], cv=5, return_train_score=False)
    clf.fit(X_train, y_train)
    scores.append({
        'model': algo,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    best_estimators[algo] = clf.best_estimator_
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

```python
best_estimators['svm'].score(X_test,y_test)
```

## Understanding the Confusion Matrix

The confusion matrix is a fundamental tool in machine learning for evaluating the performance of classification models. It presents a summary of the prediction results on a classification problem. The number of correct and incorrect predictions are summarized with count values and broken down by each class. Here's what each part of the matrix represents:

- **True Positives (TP)**: The cases in which the model correctly predicted the positive class.
- **True Negatives (TN)**: The cases in which the model correctly predicted the negative class.
- **False Positives (FP)**: The cases in which the model incorrectly predicted the positive class.
- **False Negatives (FN)**: The cases in which the model incorrectly predicted the negative class.

### Interpreting the Matrices


- **Diagonal Values**: Represent the number of correct predictions for each class (TP for each class). For example, there were 8 correct predictions for class 0, 7 for class 1, 5 for class 2, 10 for class 3, and 5 for class 4.
- **Off-diagonal Values**: Represent the number of incorrect predictions, where the model has predicted a class different from the true class (FP and FN for each class). For instance, class 0 was incorrectly predicted as class 1 once, and class 2 was predicted as class 1 five times.

The confusion matrix is a powerful diagnostic tool for understanding not only the overall accuracy of the model but also which classes are being confused with one another, which can be particularly useful in fine-tuning the model or addressing class imbalances.
